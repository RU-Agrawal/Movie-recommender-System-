```python
#Applying min max normalization in rating column to convert rating between 1 to 5
df.rename(columns = {'vote_average':'Rating'}, inplace = True) 
minrating=min(df['Rating'])
maxrating=max(df['Rating'])
newmin=1
newmax=5
for i in range(5):
    Rdas=((df['Rating'][i]-minrating)/(maxrating-minrating)*(newmax-newmin)+newmin)
    print("before",df['Rating'][i])
    df.replace(df['Rating'][i],value=Rdas,inplace=True)
    print("after",df['Rating'][i])
```

```python
column_names=['user_id','item_id','ratings','timestamp']
```

```python
df1=pd.read_csv('u.data',sep='\t',names=column_names) #'\t' for tab seperated values
df1.head()
```

```python
movie_titles=pd.read_csv('Movie_Id_Titles')
```

```python
movie_titles.head()
```

```python
df1=pd.merge(df1,movie_titles,on='item_id')
```

```python
df1.head()
```

```python
sns.set_style('white')
```

```python
#to get the average rating
df1.groupby('title')['ratings'].mean().sort_values(ascending=False).head()
```

```python
#movies with the most number of the ratings
df1.groupby('title')['ratings'].count().sort_values(ascending=False).head()
```

```python
rating=pd.DataFrame(df1.groupby('title')['ratings'].mean())
```

```python
rating.head()
```

```python
#creating number of ratings column after the ratings column
rating['num of ratings']=pd.DataFrame(df1.groupby('title')['ratings'].count())
```

```python
rating.head()
```

```python
#EDA
rating['num of ratings'].hist(bins=50)
```

```python
rating['ratings'].hist(bins=70)
```

```python
#to shoe the relationship between actual ratings and number of ratings
sns.jointplot(x='ratings',y='num of ratings',data=rating,alpha=0.5)
```

```python
#creating a matrix which has user id at one axis and movie id at the other
moviemat=df1.pivot_table(index='user_id',columns='title',values='ratings')
```

```python
moviemat.head()
```

```python
rating.sort_values('num of ratings',ascending=False).head(10)
```

```python
#Grabbing the user ratings for Star wars nad Liar Liar
starwars_user_ratings=moviemat['Star Wars (1977)']
liarliar_user_ratings=moviemat['Liar Liar (1997)']
```

```python
starwars_user_ratings.head()
```

```python
#getting similar movies to Starwars
similar_to_starwars=moviemat.corrwith(starwars_user_ratings)
similar_to_starwars
```

```python
similar_to_liarliar=moviemat.corrwith(liarliar_user_ratings)
similar_to_liarliar
```

```python
#dropping the Null Values
corr_starwars=pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars .dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
#If we want to set some min number of ratings values
corr_starwars=corr_starwars.join(rating['num of ratings'])
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
